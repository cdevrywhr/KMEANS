```r
#install packages
install.packages("factoextra")
```

```r
#import library 
library(factoextra)
```

```r
#membuat variabel data df yg menampung data mtcars (sdh tersedia pada R)
df <- mtcars
```

```r
#menghilangkan baris data yg mengandung nilai NA
df <- na.omit(df)
```

**Implementasi KMeans**
**kmeans(x, centers, nstrart) **

```r
km <- kmeans(df,centers=4,nstart=25)
```

```r
#memvisulaisasikan hasil dari clustering
fviz_cluster(km,data=df)
```

```r
#percobaan centers(jumlah cluster) = 5
km <- kmeans(df,centers=5,nstart=25)
```

```r
fviz_cluster(km,data=df)
```
